```python
def accuracy_2d(cm): 
    tn, fp, fn, tp = cm.ravel()
    falses=fp+fn
    trues = tp+tn
    return 1/(1 + (falses/trues))
```

```python
def precision_2d(cm): 
    tn, fp, fn, tp = cm.ravel()
    return 1/(1 + (fp/tp))
```

```python
def accuracy_3d(cm): 
    t0, f1_0, f2_0, f0_1, t1, f2_1, f0_2, f1_2, t2 = cm.ravel()
    falses = f1_0 + f2_0 + f0_1 + f2_1 + f0_2 + f1_2
    trues = t0 + t1 + t2
    return 1/(1 + (falses/trues))
```

```python
# taking 'positive' class to be class 1 for comparison's sake
def precision_3d(cm): 
    t0, f1_0, f2_0, f0_1, t1, f2_1, f0_2, f1_2, t2 = cm.ravel()
    f1 = f1_0 + f1_2
    return 1/(1 + (f1/t1))
```
